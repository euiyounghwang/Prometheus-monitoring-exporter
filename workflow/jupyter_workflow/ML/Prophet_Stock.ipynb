```python
df = pd.read_csv('./AAPL.csv')
df.head()
```

```python
df = df.reset_index()
df['ds'] = df['Date']
df['y'] = df['Close']
data = df[['ds','y']]
data = data[-1000:]

data.head()
```

```python
prophet_m = Prophet().fit(data)
```

```python
future = prophet_m.make_future_dataframe(periods=365, freq = 'D') # period에 할당된 날짜만큼 미래 날짜 만들기

# 미래 가격 예측하기
forecast = prophet_m.predict(future)
# forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
future.tail()
```

```python
m = prophet_m.plot(forecast)
```
